## Season 1

```python
# Getting rid of the empty str

for line in temp:
    if line == '':
        temp.remove(line)
```

```python
# Merging str with less than 2 words to the previous sent
for line in enumerate(temp):
    if len(line[1].split()) < 2:
        temp[line[0]-1] = temp[line[0]-1] + ' ' + line[1]
        
        
```

```python
# Getting rid of the str with less than 2
new_temp = []
for line in temp:
    if len(line.split()) > 2:
        new_temp.append(line)
        
new_temp
```

```python
# Removing irrelevant characters txt

for line in new_temp:
    if (":" in line.split()) and (line.split()[0] not in characters):
        new_temp.remove(line)
```

Most common words for seasons 1, 2, 3 and 5 combined

```python
# Concatenate lists of words spoken in a given season to one another
all_seasons = c['season1'] + c['season2'] + c['season3'] + c['season5']

all_s_count = (word for word in all_seasons)

all_seasons_words = Counter(all_s_count)
#all_seasons_words.most_common(200)
```
